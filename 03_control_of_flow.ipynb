### while문
- while은 어떤 조건이 참일 때 명령을 계속 반복하는 용도
- 반복문의 제어(break, continue)

### if문
- if문은 조건에 따라 다른 명령을 실행하는 용도

### for문
- for문은 컨테이너형 변수의 요소를 하나씩 순서대로 가져와 실행하는 용도의 반복문
    - range() 특정 범위의 숫자 list를 생성, for문과 결합해서 많이 사용

````python
for i in range(1,10+1):
    print(i)
    
for i in range(1,20+1):
    if i % 3 == 0:
        print('짝')
    else:
        print(i)
        
for i in range(2, 10):
    for j in range(1, 10):
        print(f'{i} x {j} = {i*j}')
````

### 오류 예외(try-except)
- 오류가 발생했을 때 예외처리를 해서 코드 실행이 멈추지 않게 할 수 있다.
```python
try:
    x= 10 / 0
except Exception as e:
    print(f'오류:{e}')
```

````python
# BMI 계산하는 프로그램에서 키를 100에서 200 사이의 값을 입력하지 않은 경우 입력을 다시 받도록 만드시오

while True:
    w = float(input())
    h = float(input())
    if h < 100 or h > 200:
        print('키는 100~200 값이어야 합니다.')
        continue
    else:
        break
    
bmi = w / (h/100)**2
print(bmi)

# 숫자가 아닌 값을 입력해 오류가 발생한 경우 다시 입력 받도록 만드시오
while True:
    try:
        w = float(input())
        h = float(input())
    except:
        print('숫자를 입력하세요')  
        continue
    if h < 100 or h > 200:
        print('키는 100~200 값이어야 합니다.')
        continue
    else:
        break
    
bmi = w / (h/100)**2
print(bmi)
````

````python
'''
과제3
1) 정수형 숫자 A,B를 input()으로 입력받아 크기를 비교한 뒤 다음과 같이 출력하시오
    - A가 B보다 10 이상 크면 A가 B보다 많이 큽니다.
    - A가 B보다 1~9 크면 A가 B보다 약간 큽니다
    - A와 B가 같으면 A가 B랑 같습니다
    - A가 B보다 1~9 작으면 A가 B약간 작습니다
    - A가 B보다 10이상 작으면 A가 B보다 많이 작습니다
    숫자가 아닌 경우 비교가 불가능합니다
'''
# 숫자가 아닌 경우를 받기 위해 try except 구문 사용
try: 
    # A, B 입력 받기
    A = int(input('정수 A를 입력하세요: '))
    B = int(input('정수 B를 입력하세요: '))
    
    # 조건 설정
    if A - B >= 10:
        print("A가 B보다 많이 큽니다.")
        
    elif 1 <= A - B <= 9:
        print("A가 B보다 약간 큽니다.")
        
    elif A == B:
        print("A가 B랑 같습니다.")
        
    elif -9 <= A - B <= -1:
        print("A가 B보다 약간 작습니다.")
        
    elif A - B <= -10:
        print("A가 B보다 많이 작습니다.")

# 숫자가 아닌경우
except ValueError:
    print("비교가 불가능합니다.")

'''
2) 주사위 2개를 던져 나올 수 있는 조합의 경우의 수를 모두 출력하시오
단, (1,2)와 (2,1) 처럼 순서가 다르더라도 같은 숫자로 이루어진 조합은 동일한 조합으로 간주해 한 가지만 출력한다.
'''
# 주사위 입력
dice = [1, 2, 3, 4, 5, 6]

# 정답 리스트 생성
ans = []

# 중복이 나오지 않도록 조건 설정
for i in dice:
    for j in dice:
        if i <= j:
            ans.append((i,j))

# 정답 출력
for i in ans:
    print(i)
````
