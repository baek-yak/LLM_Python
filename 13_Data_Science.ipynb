````python
'''
예측모델 실습
데이터를 train, test로 갈라
train 데이터를 이용해 model fitting
test 데이터를 이용해 성능 측정

- 사용 예측 모델
    - logistic regression : 선형성에 기반한 이항형 문제를 예측하는 모델
    - Random forest : 기준값에 따라 의사결정을 하는 형태의 결정트리를 다수 훈련시켜 예측의 최빈값 또는 평균값을 이용해 예측하는 모델
'''
df = pd.read_csv('data.csv')
# 데이터 크기 확인
print(df.shape)

# 필요없는 데이터 제거
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)

# M, B를 숫자로 변환
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
print(df.shape)
````

````python
# 특징 선택
from sklearn.model_selection import train_test_split

# 7:3으로 데이터를 갈라 사용
train, test = train_test_split(df, test_size=0.3)

# 특징 10개만 상관계수 구하기
corr = df.iloc[:, 1:11].corr()

# 상관관계 시각화
sns.heatmap(corr, annot=True, cmap="coolwarm")


# 임의의 네가지 특징만 이용
features = ['radius_mean', 'texture_mean', 'smoothness_mean', 'compactness_mean']
````

````python
# train과 test, x와 y로 분리 
train_X = train[features]
train_y = train['diagnosis']
test_X = test[features]
test_y = test['diagnosis']
````

````python
# 로지스틱 회귀
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# 로지스틱 회귀 객체를 만들어서
model1 = LogisticRegression()

# train 데이터를 이용해 학습
model1.fit(train_X, train_y)

# 학습된 모델의 계수 출력
print(model1.intercept_, model1.coef_)

# 검증데이터로 성능 측정
# 예측
pred1 = model1.predict(test_X)
#혼동행렬 출력
print(metrics.confusion_matrix(test_y, pred1))
#성능지표 출력
print(metrics.classification_report(test_y, pred1))
'''
precision 정밀도 : 예측한 것 중에서 실제로 맞춘 비율
recall 재현율 : 실제 답에서 모델이 정확히 예측한 비율
f1-score : 정밀도와 재현율의 조화 평균
'''
````

````python
# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier

# 랜덤포레스트 객체
model2 = RandomForestClassifier()

# train 데이터를 이용해 학습
model2.fit(train_X, train_y)

# 검증데이터로 성능 측정
# 예측
pred2 = model2.predict(test_X)
#혼동행렬 출력
print(metrics.confusion_matrix(test_y, pred2))
#성능지표 출력
print(metrics.classification_report(test_y, pred2))
'''
precision 정밀도 : 예측한 것 중에서 실제로 맞춘 비율
recall 재현율 : 실제 답에서 모델이 정확히 예측한 비율
f1-score : 정밀도와 재현율의 조화 평균
'''
````
